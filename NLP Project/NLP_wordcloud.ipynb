```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
```

```python
texts=('Make in India is a visionary initiative launched by the Indian government with the aim of transforming the country into a global manufacturing hub. This ambitious campaign seeks to boost domestic production, attract foreign investments, and create job opportunities across various sectors. By promoting ease of doing business, simplifying regulations, and investing in infrastructure, the initiative encourages both local and international businesses to set up manufacturing units in India. The campaign focuses on key sectors such as electronics, automobiles, defense, and renewable energy, aiming to enhance self-reliance and reduce dependence on imports. Make in India not only emphasizes economic growth but also strives to position India as a competitive player in the global market, fostering innovation, skill development, and sustainable manufacturing practices. As the initiative continues to evolve it plays a pivotal role in shaping Indias economic landscape and contributing to its emergence as a major manufacturing destination on the world stage.')
```

```python
texts
```

```python
wordcloud=WordCloud(width=400,height=400,margin=0,mode='RGBA', prefer_horizontal=0.9,).generate(texts)
```

```python
wordcloud
```

```python
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```
